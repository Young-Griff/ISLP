### Printing to a Stream
The **print()** method in Python can be used to print strings to the terminal or to output files. This is a very basic and commonly used function. But, here we discuss additional keyword arguments that can be passed to **print()** to get a desired formatting. In particular:
- **sep**: Defines the separator used to combine comma separated args passed to the function.
- **end**: Specifies the last character output by the function. By default this is a newline character '\n'.
- **file**: An object with a write method dictating where the output string should be printed. If not specified, sys.stdout is used.

```python
# Print a set of strings separated by hyphens '-'
print("Clubs", "Diamonds", "Hearts", "Spades", sep=' - ')
# Print a sentence followed by a line of asterisks
endline = "\n" + ("*" * 39) + "\n"
print("Those are the suits in a deck of cards!", end=endline)
# Write a string to a file
with open('hello_world.txt', 'w') as f:
    print("Hello, World!", file=f)
```

```python
with open('hello_world.txt', 'r') as f:
    print(f.readline(), end="")
```

### Getting Function Help
Sometimes, when working with a function it may be useful to know more about the function. What arguments does it take? What does it return? This is where the **help()** function is useful. This function takes another function as its argument and returns the docstring for the passed function.

```python
# Get the docstring for print
help(print)
```

Docstrings are triple quoted strings at the start of a function. For example, a doctstring for an add function could be:

```python
def add(a: int, b: int) -> int:
    """
        Adds two integers together and returns the resulting integer.
    """
    return a + b

help(add)
```

### Interating Over Ranges
Python comes with a built in **range()** function that can be used to iterate over predefined ranges of numbers. The function takes three positional arguments with only the second being required. These are:
- **start**: The number with which to begin the range (default is 0).
- **end**: The number at which to stop the range (only required argument and not inclusive).
- **step**: An integer specifying the increment (default is 1).

The produced range includes all integers in $[start, end)$ at the specified increment. 

```python
# Get numbers from 0 - 5
stop = 5
[i for i in range(stop + 1)]
```

Notice, in the case above, because the value of stop is desired in the output, one is added to the value of stop in the **range()** function.

```python
# Get numbers from 25 - 35
start = 25
end = 35
print([i for i in range(start, end + 1)])
step = 2
# Repeat but step by 2
print([i for i in range(start, end + 1, step)])
```

An invalid range will result in an empty iterable.

```python
[i for i in range(10, 20, -1)]
```

### Apply a Function Across an Iterable
A function can be applied over an iterable explicitly in a for loop or, more succinctly, using the **map()** function. The **map()** function takes two positional arguments:
1. A function to be applied to each element.
2. An iterable object over which to apply the function.

The passed function can be defined outside of the **map()** function or can be defined in-place as a lambda function.

```python
my_list = list(range(6))
# Get the square of each number using a pre-defined function
def square(x):
    return x**2
print(list(map(square, my_list)))
# Perform the same operation w/ a lambda function
print(list(map(lambda x: x**2, my_list)))
```

### Filtering an Interable
Filtering (i.e. selecting certain elements of an iterable meeting a given criteria) is a common problem. Python provides a built-in **filter()** function that can be used to filter. The function takes two arguments:
1. A function that returns a boolean indicating if an element satisfies the filter criteria.
2. The iterable over which to apply the filter.

Like with the **map()** function, the **filter()** function's provided function can be pre-defined or can be done in-place as a lambda function.

```python
my_strs = ["Apple", "arm", "nvda", "", "AMD"]
# Filter w/ pre-defined function
def a_prefix(s):
    return (len(s) > 0) and (s[0].lower() == 'a')
print(list(filter(a_prefix, my_strs)))
# FIlter w/ lambda function
lambda_filter = filter(
    lambda s: (len(s) > 0 and s[0].lower() == 'a'),
    my_strs
)
print(list(lambda_filter))
```

### Sorting an Iterable
The **sorted()** function is a commonly used built-in Python function. The function takes an iterable as a positional argument which it sorts in ascending order. Several key word arguments can be included to alter the result:
1. The boolean **reverse** can be set to True to sort the iterable in descending order.
2. A **key** function can be provided to specify which value to use for sorting when sorting more complex data structures.

```python
# Perform default and reversed sorting
print(sorted(my_strs))
print(sorted(my_strs, reverse=True))
# Perform sorting on a list of dictionaries (key must be specified)
my_dicts = [{"name": "Tom", "age": 27}, {"name": "Bill", "age": 92}, {"name": "Joe", "age": 26}]
print(sorted(my_dicts, key= lambda d: d["name"]))
```

### Enumerating Iterable Components
In coding, it is common practice to use a for loop to iterate through an array using an index variable incremented at each step. This index can then be used for accessing the corresponding iterable element and for other operations within the loop. Python's built-in **enumerate()** function allows for iterating through a loop without the need for indexing to retrieve the current element. It does so by returning a tuple with each element and its index. The **enumerate()** function simply takes an iterable object as its argument.

```python
show_enum = ["This", "is", "enumerate"]
print(list(enumerate(show_enum)))
```

### Combining Iterables
Sometimes, it may be desireable to combine information from two or more iterables. This process can be more complicated than it seems (handling varying numbers of arrays and varying lengths can be tricky to do cleanly). Python's built-in **zip()** function can handle this process for users. The function takes a variable number of iterables as positional arguments and combines their components into tuples in the order provided. If the iterables are of unequal length, the resulting iterable will only be as long as the shortest provided.

```python
a = range(11)
b = range(10, -1, -1)
c = range(-5, 1)
print("Equal:", list(zip(a, b)))
print("Unequal:", list(zip(c, b, a)))
```
