## Indexing Data
In this section, a two-dimensional array (matrix) is used to demonstrate indexing. Indexing is the process of accessing elements in a list or array. As discussed earlier in this lab, a **numpy** array can be indexed using the format **arr[row, col]**. Below, the deomstration matrix is created and displayed along with the third element of the second row (remember **numpy** arrays are 0 indexed).

```python
A = np.array(np.arange(16)).reshape((4, 4))
A, A[1, 2]
```

To select only certain rows or columns, a list of the desired indices is passed in either the **row** or **col** positions. A colon (":") is used to select all rows or columns. For example, to get the second element of the first and second rows or to get the second and fourth elements of the second row the following code can be executed.

```python
A[[0, 2], 1], A[1, [1, 3]]
```

Unfortunately, this process cannot be used to extract submatrices with multiple rows and columns without adjustment.

```python
A[[1,3], [3,2]]
```

```python
A[[1,3],[1,2,3]]
```

As seen above, attempting to extract a pair of rows and columns results in **numpy** assuming the user is requesting two points ([1, 3] and [3, 2]) when the goal was truly to extract the fourth and third elements of the second and fourth row. Additionally, it can be seen that **numpy** expects the size of lists used for indexing to match (this is because, as discussed earlier, **numpy** interprets multiple passed lists to be coordinates of elements in the matrix). Luckily, there are simple methods of working around this issue. 
1. A simple method is to first extract the desired rows then extract the desired columns from the extracted rows (ex. A[[1,3]][:,[0,4]])
2. A more efficient means to extract a submatrix is to use the **np.ix_(rows, cols)** fuction which creates a submatrix by using an intermediate mesh object.
3. Slices can also be used to extract submatrices using using the format **low**:**high**:**step**. For example, [0:3] defaults to steps of 1 and extracts elements 0, 1, and 2 while [0:3:2] uses steps of 2 and extracts the elements 0 and 2.
The codeblock below demonstrates that all 3 methods can be used to achieve the same result.

```python
idx = np.ix_([1,3], [1, 2, 3])
A[idx], A[[1, 3]][:, [1, 2, 3]], A[1:4:2, 1:4]
```

### Boolean Indexing
In **numpy**, a *boolean* is a type that can be either **True** or **False** (can also be represented by 1 or 0, respectively). Below, **np.zeros()** is used to create an array of zeros interpeted as booleans. 

```python
b = np.zeros(5, bool)
b
```

As with standard **numpy** arrays, elements of *boolean* arrays can be changed. 

```python
b[[1, 3]] = True
b
```

As discussed above, booleans can be interpreted as numerical values. As such, the following is true:

```python
c = np.array([0, 1, 0, 1, 0])
np.all(b == c)
```

Though these two arrays are considered equal, they do not return the same results when used to index an matrix. The array **c** will return the first and second rows in an alternating manner. The array **b** will return the rows where the array element is **True**.

```python
B = np.linspace(0, 28, 15).reshape((5, 3))
B
```

```python
B[b], B[c]
```

Boolean arrays can be used with the **np.ix_()** to create a submatrix. The function will retain rows and columns where the value is the passed boolean array is **True**.

```python
rows = np.zeros(5, bool)
cols = np.zeros(3, bool)
rows[[0, 2, 4]] = True
cols[1] = True
idx = np.ix_(rows, cols)
B[idx]
```

Note that a list of integers can be mixed with a list of booleans in the arguments for **np.ix_()**.

```python
idx = np.ix_([0, 2, 4], cols)
B[idx]
```
